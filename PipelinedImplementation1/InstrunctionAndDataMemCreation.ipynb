```python
from random import seed
from random import randint
from bitstring import Bits
```

```python
regDict = {
    '$zero' : '00000',
    '$at' : '00001',
    '$v0' : '00010',
    '$v1' : '00011',
    '$a0' : '00100',
    '$a1' : '00101',
    '$a2' : '00110',
    '$a3' : '00111',
    '$t0' : '01000',
    '$t1' : '01001',
    '$t2' : '01010',
    '$t3' : '01011',
    '$t4' : '01100',
    '$t5' : '01101',
    '$t6' : '01110',
    '$t7' : '01111',
    '$s0' : '10000',
    '$s1' : '10001',
    '$s2' : '10010',
    '$s3' : '10011',
    '$s4' : '10100',
    '$s5' : '10101',
    '$s6' : '10110',
    '$s7' : '10111',
    '$t8' : '11000',
    '$t9' : '11001',
    '$k0' : '11010',
    '$k1' : '11011',
    '$gp' : '11100',
    '$fp' : '11101',
    '$sp' : '11110',
    '$ra' : '11111'
}
opCode = {
    'lw' : '100011',
    'sw' : '101011',
    'add' : '000000',
    'sub' : '000000',
    'beq' : '000100',
    'j' : '000010'
}
funct = {
    'add' : '100000',
    'sub' : '100010'

}
```

```python
def InstructionCreater(instructionWithSpace):
    print(instructionWithSpace)
    spaceSeperated = instructionWithSpace.split(' ')
    instruction = spaceSeperated[0]
    if instruction in ['lw', 'sw', 'beq']: # for i formate
        rt = spaceSeperated[1][:-1]
        if instruction in ['beq']:
            rs  = spaceSeperated[2][:-1]
            cons = Bits(int=int(spaceSeperated[3]), length=16).bin
        else:
            bractetSplit = spaceSeperated[2].split('(')
            cons = Bits(int=int(bractetSplit[0])*4, length=16).bin
            rs = bractetSplit[1][:-1]        
        ins = hex(int('0b'+opCode[instruction]+regDict[rs]+regDict[rt]+cons,2))[2:] 
    elif instruction in ['add', 'sub']: # for R format
        rd = spaceSeperated[1][:-1]
        rs = spaceSeperated[2][:-1]
        rt = spaceSeperated[3]
        ins = hex(int('0b'+opCode[instruction]+regDict[rs]+regDict[rt]+regDict[rd]+'00000'+funct[instruction], 2))[2:]
    elif instruction in ['j']: # for J format
        addr = Bits(int=int(spaceSeperated[1]), length=25).bin        
        ins = hex(int('0b'+opCode[instruction]+addr,2))[2:]


    return ins
# InstructionCreater('sw $t5, 11($zero)')
# InstructionCreater('sub $s4, $t8, $t9')
# InstructionCreater('beq $zero, $zero, 7')
# InstructionCreater('j 15')
```

```python
def writeDataMemory(depth=10, initwithzero=1, randomize=0):
    if initwithzero:
        st='00000000\n'*depth
    else:
        st = ''
        if randomize:
            seed(105)
            lis = [hex(randint(0, 40))[2:] for i in range(depth)]
            st = '\n'.join(lis)
    with open('MemFiles/DataMemory.mem', 'w+') as f:
        f.write(st)
        f.close()
def writeInstructionMemory():
    st = ''
    # ins 1 to 10
    for i in range(10):
        ins = InstructionCreater('lw $t'+str(i)+', '+str(i)+'($zero)')
        # ins = hex(int('0b'+opCode['lw']+regDict['$zero']+regDict['$t'+str(i)]+bin(i*4)[2:].zfill(16),2))[2:]    
        st =  st + ins + '\n'

    # ins 11
    # SEE currentPC value - jumps to 15 th instruction
    # ins1 = InstructionCreater('j 15')
    # ins1 = hex(int('0b'+opCode['j']+bin(15)[2:].zfill(26),2))[2:]
    # st = st + ins1+'\n'

    # ins 12
    # ins1 = hex(int('0b'+opCode['add']+regDict['$t0']+regDict['$t1']+regDict['$s1']+'00000'+funct['add'], 2))[2:]
    ins1 = InstructionCreater('add $s1, $t0, $t1')
    st = st + ins1+'\n'
    # # ins 13
    # ins1 = hex(int('0b'+opCode['add']+regDict['$t2']+regDict['$t3']+regDict['$s2']+'00000'+funct['add'], 2))[2:]
    ins1 = InstructionCreater('add $s2, $t2, $t3')
    st = st + ins1+'\n'
    # # ins 14
    # ins1 = hex(int('0b'+opCode['sub']+regDict['$t6']+regDict['$t5']+regDict['$s3']+'00000'+funct['sub'], 2))[2:]
    ins1 = InstructionCreater('sub $s3, $t6, $t5')
    st = st + ins1+'\n'
    # # ins 15
    # ins1 = hex(int('0b'+opCode['sub']+regDict['$t8']+regDict['$t9']+regDict['$s4']+'00000'+funct['sub'], 2))[2:]
    ins1 = InstructionCreater('sub $s4, $t8, $t9')
    st = st + ins1+'\n'
    # # ins 16
    # ins1 = hex(int('0b'+opCode['sw']+regDict['$zero']+regDict['$t7']+bin(0*4)[2:].zfill(16),2))[2:]  
    ins1 = InstructionCreater('sw $t7, 0($zero)')
    st = st + ins1+'\n'
    # # ins 17
    # ins1 = hex(int('0b'+opCode['sw']+regDict['$zero']+regDict['$t6']+bin(1*4)[2:].zfill(16),2))[2:]   
    ins1 = InstructionCreater('sw $t6, 1($zero)')
    st = st + ins1+'\n'
    # # ins 18
    # ins1 = hex(int('0b'+opCode['sw']+regDict['$zero']+regDict['$t5']+bin(11*4)[2:].zfill(16),2))[2:]   
    ins1 = InstructionCreater('sw $t5, 11($zero)')
    st = st + ins1+'\n'
    # # ins 19
    # # SEE currentPC value - braches back to 7 instructions to 12th
    # ins1 = hex(int('0b'+opCode['beq']+regDict['$zero']+regDict['$zero']+Bits(int=-7, length=16).bin,2))[2:]
    ins1 = InstructionCreater('beq $zero, $zero, -7')
    st = st + ins1+'\n'

    ins1 = InstructionCreater('sw $t7, 0($zero)')
    st = st + ins1+'\n'
    ins1 = InstructionCreater('sw $t6, 1($zero)')
    st = st + ins1+'\n'
    ins1 = InstructionCreater('sw $t7, 0($zero)')
    st = st + ins1+'\n'
    ins1 = InstructionCreater('sw $t6, 1($zero)')
    st = st + ins1+'\n'

    
    with open('MemFiles/InstructionMemory.mem', 'w+') as f:
        f.write(st)
        f.close()
```
